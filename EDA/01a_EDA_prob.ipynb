````python
#이게 왜 안될까? ㅋㅋ
df_list = []
for i in [df1, df2]:
    
    i = i.T
    i = i.rename(columns = i.iloc[0]).drop('주요재무정보', axis = 0)
    df_list.append(i)
    
df1 = df_list[0]
df2 = df_list[1]
df1.head()
````

````python
df1 = df1.T
df1 = df1.rename(columns = df1.iloc[0]).drop('주요재무정보', axis = 0)
df1.head()
````

````python
df2 = df2.T
df2 = df2.rename(columns = df2.iloc[0]).drop('주요재무정보', axis = 0)
df1.head()
````

이를 참조하여, `pp`의 `species`와 `sex`변수에 대해서 cross-tabulation을 시행하고, 6개의 값들 중 가장 큰 값을 `num3`에, 가장 작은 값을 `num4`에 저장하세요.

````python
import seaborn as sns

pp = sns.load_dataset('penguins')
````

````python
pp.head()
pp.shape
pp.info()
````

````python
# 교차표 생성

cross_pp = pd.crosstab(pp.species, pp.sex)
cross_pp
````

````python
num3 = cross_pp.max()
num4 = cross_pp.min()
````
